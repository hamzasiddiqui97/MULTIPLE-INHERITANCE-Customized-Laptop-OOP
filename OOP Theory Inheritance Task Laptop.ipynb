```python
class RAM:
    def __init__(self, ram_storage, ram_speed):
        self.ram_storage = ram_storage
        self.ram_speed = ram_speed
    def ram_info(self):
        print('RAM Storage: {}gb \nRAM speed: {}'.format(self.ram_storage,self.ram_speed))
        print(30*'_')

class SCREEN:
    def __init__(self,screen_hertz, screen_size, screen_aspect_ratio):
        self.screen_hertz = screen_hertz
        self.screen_size = screen_size
        self.screen_aspect_ratio = screen_aspect_ratio

    def screen_info(self):
        print('Screen hertz: {}hertz \nSize: {}inches \nAspect Ratio: {}'.format(self.screen_hertz, self.screen_size,self.screen_aspect_ratio))
        print(30*'_')

class GraphicsCARD:
    def __init__(self, gpu_name ,gpu_Vram):
        self.gpu_name = gpu_name
        self.gpu_Vram = gpu_Vram

    def gpu_info(self):
        print('GPU name: {} \nVram: {}gb'.format(self.gpu_name,self.gpu_Vram))


class CPU:
    def __init__(self,cpu_cores, cpu_speed,cpu_ram, cpu_model, cpu_processor):
        self.cpu_cores = cpu_cores
        self.cpu_speed = cpu_speed
        self.cpu_ram = cpu_ram
        self.cpu_model = cpu_model
        self.cpu_processor = cpu_processor

    def cpu_info(self):
        print('Processor Name: {} \nCore: {} \nRam: {}gb \nProcessor Speed: {}Ghz'.format(self.cpu_model,self.cpu_cores,self.cpu_ram,self.cpu_speed))
        print(30*'_')

class Laptop(RAM, SCREEN,GraphicsCARD, CPU):
    def __init__(self, laptop_brand, laptop_release_date):
        self.laptop_brand = laptop_brand
        self.laptop_release_date = laptop_release_date
        
    def laptop_details(self):
        print('Laptop Brand: {} \nRelease Date: {}'.format(self.laptop_brand, self.laptop_release_date))

    def ram_print(self,ram_storage, ram_speed):
        RAM.__init__(self, ram_storage, ram_speed)
    def screen_print(self,screen_hertz, screen_size, screen_aspect_ratio):
        SCREEN.__init__(self,screen_hertz, screen_size, screen_aspect_ratio)
    def graphics_print(self,gpu_name ,gpu_Vram):
        GraphicsCARD.__init__(self,gpu_name ,gpu_Vram)
    def cpu_print(self,cpu_cores, cpu_speed,cpu_ram, cpu_model, cpu_processor):
        CPU.__init__(self,cpu_cores, cpu_speed,cpu_ram, cpu_model, cpu_processor)
        

ob1 = Laptop('HP','12')
ob1.ram_print(8, 3)
ob1.screen_print(60,19,'1080p')
ob1.graphics_print('rx560',2)
ob1.cpu_print(2,3,8,'I5', 'I5')

ob1.ram_info()
ob1.screen_info()
ob1.cpu_info()
ob1.gpu_info()
ob1.laptop_details()
```
